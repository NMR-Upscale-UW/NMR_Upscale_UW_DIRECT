This sheets converts the individual NMR Spectra data into featurization sheets
1. Read all the files.
2. Take the Y values of 60 MHz and fill the data into column vectors
3. Take the X values of 60 MHz and create column vectors
4. Run it for all the files in the directory
5. Take the final dataframe and save it as csv
6. Do it for both 60 MHz and 400 MHz

```python
import glob
ofilenames = glob.glob("NMRfiles_Test/*.csv")
ofilenames.sort()
ofilenames
```

```python
import pandas as pd
import os
import numpy as np

data_temp=pd.read_csv(ofilenames[0])
df_60 = np.zeros(shape=(len(ofilenames), len(data_temp.index)))
df_400 = np.zeros(shape=(len(ofilenames), len(data_temp.index)))

x= data_temp["400MHz_ppm"].to_numpy()

for index, f in enumerate(ofilenames):
    data=pd.read_csv(f)
    df_60[index]=data["60MHz_intensity"].to_numpy()
    df_400[index]=data["400MHz_intensity"].to_numpy()
df60 = pd.DataFrame(df_60,columns=x)
df400 = pd.DataFrame(df_400,columns=x)
df60.to_csv('NMRFiles_Test/60MHz.csv', index=False)
df400.to_csv('NMRFiles_Test/400MHz.csv', index=False)
# For loop, read each file, get the array and append it as a row into two dataframes
```

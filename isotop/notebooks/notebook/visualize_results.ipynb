```python
import matplotlib.pyplot as plt
import torch
import numpy as np
```

```python
results = torch.load("../results/cnn/predictions.pt")
loss_curves = torch.load("../results/cnn/loss_curves.pt")
```

```python
valid_l = loss_curves['valid_loss']
train_l = loss_curves['train_loss']
plt.plot(np.arange(len(valid_l)), valid_l, label='Validation Loss', c='orange')
plt.plot(np.arange(len(train_l)), train_l, label='Train Loss', c='royalblue')
plt.title("Loss Curve")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.show()
```

```python
x= np.linspace(0, 1, 5500)
for i, (low_res, high_res_pred, high_res, loss) in enumerate(results):
    plt.plot(x, low_res[0][0], c='g', label='Low Resolution') 
    plt.plot(x, high_res[0][0], c='r', label = 'High Resolution') 
    plt.plot(x, high_res_pred[0][0].detach(), c='royalblue', label='Prediction') 
    plt.xlabel("chemical shift")
    plt.ylabel("intensity")
    plt.title(f"Test Set Prediction {i+1}")
    plt.legend()
    plt.show()
```
